```python
from ultralytics import YOLO
model = YOLO('/content/runs/detect/train/weights/best.pt')
```

```python
results = model.val(data='/content/ANPR-1/data.yaml')
```

```python
# Accessing the metrics
metrics = results.metrics

# Extract the specific metrics you need
f1_score = metrics['f1']
precision = metrics['precision']
recall = metrics['recall']
map_0_5 = metrics['map50']
map_0_95 = metrics['map']

# Print the metrics
print(f"Best F1 Score: {f1_score:.4f}")
print(f"Best Precision: {precision:.4f}")
print(f"Best Recall: {recall:.4f}")
print(f"mAP@0.5: {map_0_5:.4f}")
print(f"mAP@0.5:0.95: {map_0_95:.4f}")
```
